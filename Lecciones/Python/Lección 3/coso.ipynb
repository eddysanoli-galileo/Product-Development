```python
# Imports
# pip install pyserial
# pip install adafruit-io
import serial
import time
from Adafruit_IO import Client, Feed, RequestError

# ==================
# PUERTO SERIAL
# ==================

def sendToSerial(data):

    # Nombre del puerto a utilizar
    port_name = "COM3"

    # Se crea una conexión al puerto serial
    port_connection = serial.Serial(port = port_name, baudrate = 9600)

    # Mensaje para notificar si la conexión fue exitosa
    if port_connection.isOpen():
        print(f"Connection with {port_name} successful")
    else:
        print(f"Connection with {port_name} failed")

    # Se escribe la variable al puerto seleccionado
    port_connection.write(data)

    # Se cierra el puerto
    port_connection.close()

# ==================
# ADAFRUIT IO
# ==================

# Adafruit IO key y usuario.
ADAFRUIT_IO_KEY = "aio_sHTj09TUFVxVEOOe9qctFH1mWB59"
ADAFRUIT_IO_USERNAME = 'Jocarsoli_so_2001'

# Conexión a cliente de Adafruit
aio = Client(ADAFRUIT_IO_USERNAME, ADAFRUIT_IO_KEY)

# SLIDER 1:
# Si no existe la variable buscada, se crea
try: 
    slider_1 = aio.feeds('var00')
    print("Feed 0 - Success")
except RequestError:
    feed = Feed(name='var00')
    slider_1 = aio.create_feed(feed)

# SLIDER 2:
# Si no existe la variable buscada, se crea
try:
    slider_2 = aio.feeds('var01')
    print("Feed 1 - Success")
except RequestError: 
    feed = Feed(name='var01')
    slider_2 = aio.create_feed(feed)

# Valores previos de feed 
slider1_history = 0
slider2_history = 0
```

```python
while True:

    # Objeto que almacena la variable asociada al feed_01
    read_slider1 = aio.receive(slider_1.key)
    read_slider2 = aio.receive(slider_2.key)

    # Se lee el valor del slider y s  e convierte a int
    var00 = int(read_slider1.value)
    var01 = int(read_slider2.value)

    # Slider 1:
    # Si el valor recibido es diferente al previo, se imprime.
    if (var00 != slider1_history):
        print(f"Var 00 - Val Received: {var00}")
        sendToSerial(f"S1{str(var00)[::-1]}e".encode())
        slider1_history = var00

    # Slider 2: 
    # Si el valor recibido es diferente al previo, se imprime.
    if (var01 != slider2_history):
        print(f"Var 01 - Val Received: {var01}")
        sendToSerial(f"S2{str(var01)[::-1]}e".encode())
        slider2_history = var01
```
